```python
import pandas as pd
import requests
```

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"

results = requests.get(url)
```

```python
from bs4 import BeautifulSoup
```

```python
shop = BeautifulSoup(results.content,"lxml")
```

```python
d = shop.table.find_all("tr")
```

```python
data = pd.DataFrame(columns=["PostalCode","Borough","Neighborhood"])
```

```python
data
```

```python
for i in d:
    j = i.find_all(["td"])
    try:
        code = j[0].text
        bor = j[1].text
        nei = j[2].text[0:-1]
        if(bor=="Not assigned"):
            pass
        else:
            if(nei=="Not assigned"):
                data = data.append({"PostalCode" : code,"Borough" : bor, "Neighborhood" : bor},ignore_index=True)
            else:
                data = data.append({"PostalCode" : code,"Borough" : bor, "Neighborhood" : nei},ignore_index=True)
    except IndexError:
        pass
```

```python
data
```

```python
data.shape
```

```python
df = pd.read_csv("http://cocl.us/Geospatial_data")
```

```python
codes = [[],[]]
```

```python
for i in data["PostalCode"]:
    for j in df["Postal Code"]:
        if(i==j):
            lat = df.loc[df["Postal Code"]==i]["Latitude"].values[0]
            lng = df.loc[df["Postal Code"]==i]["Longitude"].values[0]
            codes[0].append(lat)
            codes[1].append(lng)
```

```python
data.insert(3,"Latitude",codes[0],True)
data.insert(4,"Longitude",codes[1],True)
```

```python
data
```

```python
import folium
```

```python
map = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,bor,nei in zip(data["Latitude"],data["Longitude"],data["Borough"],data["Neighborhood"]):
    label = "{},{}".format(nei,bor)
    label = folium.Popup(label,parse_html=True)
    folium.CircleMarker(
        [lat,lng],
        radius = 5,
        popup=label,
        color="blue",
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map)
    
map
    
```
